```python
import matplotlib.pyplot as plt

def readFile(file):
    data = []
    with open(file) as csv:
        data = [line.split(',') for line in csv.readlines()[1:]]
    return data

def main():
    files = ["rsel.csv","cel-rs.csv","2cel-rs.csv","cel.csv","2cel.csv"]
    spec1 = ['b','g','r','k','m']

    fig = plt.figure(figsize=(6.7,4.8),dpi=150)
    ax = fig.add_subplot(111)
    
    
    for it, file in enumerate(files):
        
        fields = []
        rows = []
        avg = []
        rselX = []
        rselY = []

        data = readFile(file)
        
        for i, row in enumerate(data):
            avg.append(0)
            #print(row)
            for j, field in enumerate(row):
                #print(field)
                if j > 1:
                    avg[i] += float(row[j])
                if j == 1:
                    rselY.append(float(row[j]))
            avg[i] = avg[i]/(len(row)-2)
            avg[i] = round(avg[i],5)
        print(len(avg))
        rselX = avg  
        ax.plot(rselY,rselX,c=spec1[it],linewidth=0.8)    
        
    ax.legend(('1-Evol-RS','1-Coev-RS','2-Coev-RS','1-Coev','2-Coev'),fontsize='large',
              framealpha=1,edgecolor='black', fancybox=False,loc='lower right')
    ax.tick_params(direction='in')
    ax.set_xlim([0, 500000])
    ax.set_ylim([0.60, 1.00])
    ax.set_xlabel('Rozegranych gier') 
    ax.set_ylabel('Odsetek wygranych gier') 
    plt.savefig('myplot.pdf')
    plt.show()
    plt.close()

    
if __name__=='__main__':
    main()
```
